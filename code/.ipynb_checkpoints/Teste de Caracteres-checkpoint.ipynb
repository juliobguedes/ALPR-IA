```python
def iou(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)

    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)

    # return the intersection over union value
    return iou
```

## Analise das Pipelines

### Equalização Padrão

```python
def pipeline(image, obj, med_kernel=(7,7), sigX=21, sigY=21, blks=3, c=0, opker=(3,3), opiter=2, show=False):
    media = cv2.GaussianBlur(image.copy(), sigmaX=sigX, sigmaY=sigY, ksize=med_kernel)
    if (show):
        plot(media)
    eql = cv2.equalizeHist(media.copy())
    if (show):
        plot(eql)
    threshold = cv2.adaptiveThreshold(eql.copy(), 255, 
                                cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, C=c, blockSize=blks)
    if (show):
        plot(threshold)
    closed = cv2.morphologyEx(threshold.copy(), cv2.MORPH_OPEN, kernel=opker, iterations=opiter)
    if (show):
        plot(closed)
    _, ctrs, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    n = 0
    copy = image.copy()
    copy = cv2.cvtColor(copy, cv2.COLOR_GRAY2BGR)
    selected_contours = []
    ious = []
    for cnt in ctrs:
        (x, y, w, h) = cv2.boundingRect(cnt)
        if (12 < h < 20 and 2 < w < 15 and 3 < x < 89):
            n += 1
            ux, uw = (x - 4, w + 7) if (w < 5) else (x,w)
            cv2.rectangle(copy, (ux,y-5), (ux+uw,y+15), (0, 255, 0), 1)
            selected_contours.append((ux, y-5, uw, y+15))
            
    for cnt in selected_contours:
        for i in range(1, 8):
            x, y, w, h = map(int, obj['resized %d' % i].split(" "))
            xc, yc, wc, hc = cnt[0], cnt[1], cnt[0] + cnt[2], cnt[1] + cnt[3]

            boxA = [x, y, x+w, y+h]
            boxB = [xc, yc, wc, hc]

            score = iou(boxA, boxB)
            if (score > 0):
                ious.append(score)
    if (show):
        plot(copy)
    final_score = np.nanmean(ious) if len(ious) != 0 else 0
    return final_score
```

```python
def pipeline_summary(pipe, data, files):
    blur_values = [(i, i) for i in range(3, 25, 2)]
    block_values = [i for i in range(3, 25, 2)]
    opk_values = [(i, i) for i in range(3, 25, 2)]
    opit_values = [i for i in range(0, 15)]
    sigx_values = [i  for i in range(0, 31, 3)]
    sigy_values = [i  for i in range(0, 31, 3)]
    c_values = [i  for i in range(0, 16, 3)]
    
    iterate = [blur_values, block_values, opk_values,
              opit_values, sigx_values, sigy_values,
              c_values]
    
    best = (-1, (0, 0, 0, 0, 0, 0, 0))
    
    s = np.prod([len(i) for i in iterate])
    ini_s = int(s)
    print('Total Combinations: {}'.format(s))
    
    for blur, block, opk, opit, sigx, sigy, c in itertools.product(*iterate):
        current_score = np.average(np.fromiter((pipe(g, get_obj(e), med_kernel=blur,
                                                     blks=block, opker=opk, opiter=opit,
                                                     sigX=sigx, sigY=sigy, c=c) for g, e in zip(data, files)), float))
        if (current_score >  best[0]):
            best = (current_score, (blur, block, opk, opit, sigx, sigy, c))
            print(best[0])
        if (best[0] > 0.9 or s == ini_s - 10**4):
            break
            
        s -= 1
        if (s % 1000 == 0): print(s)
    
    (blur_score, block_score, opk_score, opit_score, sigx_score, sigy_score, c_score) = best[1]
    print('Best Blur Kernel: {}\nBest Block Size: {}'.format(blur_score, block_score))
    print('Best Open Kernel: {}\nBest Open Iteration: {}'.format(opk_score, opit_score))
    print('Best Sigma X: {}\nBest Sigma Y: {}'.format(sigx_score, sigy_score))
    print('Best C: {}'.format(c_score))
    print('Final Score: {}%%'.format(best[0]))
    
    return best[1]
```

### Equalização com limite luminosidade

```python
def pipeline_clahe(image, obj, med_kernel=(7,7), sigX=21, sigY=21, blks=3, c=0, opker=(3,3), opiter=2, show=False):
    media = cv2.GaussianBlur(image.copy(), sigmaX=sigX, sigmaY=sigY, ksize=med_kernel)
    if (show):
        plot(media)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(15,15))
    eql = clahe.apply(media.copy())
    if (show):
        plot(eql)
    threshold = cv2.adaptiveThreshold(eql.copy(), 255, 
                                cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, C=c, blockSize=blks)
    if (show):
        plot(threshold)
    closed = cv2.morphologyEx(threshold.copy(), cv2.MORPH_OPEN, kernel=opker, iterations=opiter)
    if (show):
        plot(closed)
    _, ctrs, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    n = 0
    copy = image.copy()
    copy = cv2.cvtColor(copy, cv2.COLOR_GRAY2BGR)
    selected_contours = []
    ious = []
    
    for cnt in ctrs:
        (x, y, w, h) = cv2.boundingRect(cnt)
        if (12 < h < 20 and 2 < w < 15 and 3 < x < 89):
            n += 1
            ux, uw = (x - 4, w + 7) if (w < 6) else (x,w)
            cv2.rectangle(copy, (ux,y-5), (ux+uw,y+15), (0, 255, 0), 1)
            selected_contours.append((ux, y-5, uw, y+15))
            
    for cnt in selected_contours:
        for i in range(1, 8):
            x, y, w, h = map(int, obj['resized %d' % i].split(" "))
            xc, yc, wc, hc = cnt[0], cnt[1], cnt[0] + cnt[2], cnt[1] + cnt[3]

            boxA = [x, y, x+w, y+h]
            boxB = [xc, yc, wc, hc]

            score = iou(boxA, boxB)
            if (score > 0):
                ious.append(score)
                
    if (show):
        plot(copy)
    return np.nanmean(ious) if len(ious) > 0 else 0
```
